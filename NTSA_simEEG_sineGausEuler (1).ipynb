```python
import numpy as np
import matplotlib.pyplot as plt

# 
freq  = 2    #  (Hz)
srate = 1000 #  
time  = np.arange(-1, 1+1/srate, 1/srate) #  
ampl  = 2    # 
phas  = np.pi/3  # 

#   
sinewave = ampl * np.sin(2 * np.pi * freq * time + phas)

#   
plt.figure(figsize=(8,4))
plt.plot(time, sinewave, 'k')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('My first sine wave plot! Mom will be so proud!')
plt.xlim([-1.1, 1.1])
plt.ylim([-2.2, 2.2])
plt.grid()
plt.show()
```

```python
#    
frex = np.array([3, 10, 5, 15, 35])  #  
amplit = np.array([5, 15, 10, 5, 7]) #  
phases = np.array([np.pi/7, np.pi/8, np.pi, np.pi/2, -np.pi/4]) # 

#   
sine_waves = np.array([a * np.sin(2*np.pi*f*time + p) for f, a, p in zip(frex, amplit, phases)])

#   
sum_sine = np.sum(sine_waves, axis=0)

#   
plt.figure(figsize=(8,4))
plt.plot(time, sum_sine, 'k')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sum of Sine Waves')
plt.show()

#     
plt.figure(figsize=(8,6))
for i in range(len(frex)):
    plt.subplot(len(frex), 1, i+1)
    plt.plot(time, sine_waves[i], 'k')
    plt.ylim(-max(amplit), max(amplit))
    plt.ylabel(f'F = {frex[i]} Hz')
plt.xlabel('Time (s)')
plt.tight_layout()
plt.show()
```

```python
# 
time  = np.arange(-2, 2+1/1000, 1/1000)
ptime = 1
ampl  = 45
fwhm  = 0.9

#  
gwin = np.exp(-(4*np.log(2)*(time-ptime)**2) / fwhm**2)

#   
plt.figure(figsize=(8,4))
plt.plot(time, gwin, 'k')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Gaussian Window')
plt.show()
```

```python
M = 2.4
k = 3*np.pi/4
meik = M * np.exp(1j*k)

#      
plt.figure(figsize=(10,4))

#  
plt.subplot(121, projection='polar')
plt.polar([0, k], [0, M], 'r')
plt.polar([k], [M], 'ro')
plt.title('Polar Plane')

#    ()
plt.subplot(122)
plt.plot(meik.real, meik.imag, 'ro')
plt.axhline(0, color='k', linestyle='--')
plt.axvline(0, color='k', linestyle='--')
plt.xlim(-M, M)
plt.ylim(-M, M)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.grid()
plt.title('Cartesian Plane')

plt.tight_layout()
plt.show()
```
